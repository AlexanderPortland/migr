```python
import pandas as pd

df = pd.read_csv('sep/migr_z.csv', dtype={'citizen': 'string', 'sex': 'string', 'age': 'string', 'decision': 'string', 'geo': 'string', 'TIME_PERIOD': int, 'OBS_VALUE': int, 'OBS_FLAG': 'string'})


df = df.drop("OBS_FLAG", axis=1)
df
```

```python
import numpy as np
from matplotlib import pylab as plt
plt.style.use("bmh")
plt.figure(figsize=(10,6))

df['TIME_PERIOD'].plot.hist(bins = 14) 
                        # bins = df['age'].nunique()
plt.xlabel('year')
plt.ylabel('count')
plt.title('measure of asylum applications over time')
plt.show()
```

```python
rej = df[df["decision"] == "REJECTED"]
pos = df[df["decision"] == "TOTAL_POS"]

df_copy = df.copy()

#shorten for testing
rej = rej.head()
pos = pos.head()
pos
```

```python
df2 = pd.DataFrame(columns = ['citizen', 'sex', 'age', 'geo', 'year', 'total_pos', 'rejected'])
#df_copy = df.copy()

def find_others_val(cit, sex, age, dec, geo, time, df):
    if (dec == "TOTAL_POS"):
        o_dec = "REJECTED"
    elif (dec == "REJECTED"):
        o_dec = "TOTAL_POS"

    a = df[(df['citizen'] == cit) & (df["sex"] == sex) & (df["age"] == age) & (df["decision"] == o_dec) & (df["geo"] == geo) & (df["TIME_PERIOD"] == time)]
    if (len(a.index) == 0):
        #print(0)
        return 0
    else:
        v = a['OBS_VALUE'].iloc[0]
        #print(a)
        #print(v)
        return v

def make_new_row(df_row, df):
    cit = df["citizen"][df_row]
    sex = df["sex"][df_row]
    age = df["age"][df_row]
    dec = df["decision"][df_row]
    geo = df["geo"][df_row]
    time = df["TIME_PERIOD"][df_row]
    if (dec == "TOTAL_POS"):
        #first is positive
        pos = df["OBS_VALUE"][df_row]
        rej = find_others_val(cit, sex, age, dec, geo, time, df)
        df2.loc[len(df2.index)] = [cit, sex, age, geo, time, pos, rej]
    elif (dec == "REJECTED"):
        #first is rejected
        rej = df["OBS_VALUE"][df_row]
        pos = find_others_val(cit, sex, age, dec, geo, time, df)
        df2.loc[len(df2.index)] = [cit, sex, age, geo, time, pos, rej]
    else:
        return 0
    
    
    #print(time)
    
for i in range(0, len(df.index)):
    make_new_row(i, df_copy)
    df = df.drop(i)
    if ((i % 100) == 0): 
        print(str(i) + "/" + str(len(df.index)) + "\t\t" + str(i*100/(i+len(df.index))) + "%")
    
df = df_copy

df2.to_csv('sep/merge.csv')

#make_new_row(2)
df2
```
